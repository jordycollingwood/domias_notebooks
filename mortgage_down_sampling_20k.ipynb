```python
#creating target vaart
proportions = df['action_taken_name'].value_counts(normalize=True).tolist()
proportions
```

```python
unique_values = df['action_taken_name'].unique().tolist()
unique_values
```

```python
def resample_data(df, target_column, unique_values, proportions, random_state=None, num_rows=None):
    np.random.seed(random_state)
    
    # Calculate the number of rows to sample for each unique value based on the desired proportions
    sample_sizes = [int(proportion * num_rows) for proportion in proportions]
    
    # Create an empty DataFrame to store the resampled data
    resampled_df = pd.DataFrame(columns=df.columns)
    
    # Iterate over each unique value and sample rows accordingly
    for value, size in zip(unique_values, sample_sizes):
        # Filter the original DataFrame for the current unique value
        subset = df[df[target_column] == value]
        
        # Sample rows randomly with replacement from the subset
        resampled_subset = subset.sample(n=size, replace=True, random_state=random_state)
        
        # Append the resampled subset to the resampled DataFrame
        resampled_df = resampled_df.append(resampled_subset)
    
    # Reset the index of the resampled DataFrame
    resampled_df.reset_index(drop=True, inplace=True)
    
    return resampled_df
```

```python
resampled_df = resample_data(df, 'action_taken_name', unique_values, proportions, random_state=0, num_rows=10000)
resampled_df
```

```python
resampled_df_proportions = resampled_df['action_taken_name'].value_counts(normalize=True).tolist()
resampled_df_proportions
```

```python
#eidted version of function so that it samples atleast 1 of each target variable
def resample_data_edit(df, target_column, unique_values, proportions, random_state=None, num_rows=None):
    np.random.seed(random_state)
    
    # Calculate the number of rows to sample for each unique value based on the desired proportions
    sample_sizes = [int(proportion * num_rows) if proportion * num_rows >= 1 else 1 for proportion in proportions]
    
    # Create an empty DataFrame to store the resampled data
    resampled_df = pd.DataFrame(columns=df.columns)
    
    # Iterate over each unique value and sample rows accordingly
    for value, size in zip(unique_values, sample_sizes):
        # Filter the original DataFrame for the current unique value
        subset = df[df[target_column] == value]
        
        # Sample rows randomly with replacement from the subset
        resampled_subset = subset.sample(n=size, replace=True, random_state=random_state)
        
        # Append the resampled subset to the resampled DataFrame
        resampled_df = resampled_df.append(resampled_subset)
    
    # Reset the index of the resampled DataFrame
    resampled_df.reset_index(drop=True, inplace=True)
    
    return resampled_df
```

```python
resampled_df_edit = resample_data_edit(df, 'action_taken_name', unique_values, proportions, random_state=0, num_rows=20000)
resampled_df_edit
```

```python
resampled_df_edit_proportions = resampled_df_edit['action_taken_name'].value_counts(normalize=True).tolist()
resampled_df_edit_proportions
```
