```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import stats
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#Import hdma dataset
df_uci = pd.read_csv(r'C:\Users\jordy\OneDrive\MSc_Python\Individual_Project\Data\UCI_Credit_Card.csv', index_col=0)
df_uci
```

## TSNE Perplexity 30

```python

# Drop the target column
features = df_uci.drop(columns=['default.payment.next.month'])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform t-SNE
tsne = TSNE(n_components=2, random_state=42)

# Fit and transform the scaled features with a progress bar
tsne_result = []
with tqdm(total=len(scaled_features), desc="t-SNE Progress") as pbar:
    for i in range(0, len(scaled_features), 100):  # Batch processing for better performance
        batch_result = tsne.fit_transform(scaled_features[i:i+100])
        tsne_result.append(batch_result)
        pbar.update(len(batch_result))

# Concatenate the t-SNE results
tsne_result = np.concatenate(tsne_result, axis=0)

# Create a new DataFrame with the t-SNE results
tsne_df = pd.DataFrame(data=tsne_result, columns=['Dimension 1', 'Dimension 2'])

# Plot the t-SNE scatter plot
plt.scatter(tsne_df['Dimension 1'], tsne_df['Dimension 2'])
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('t-SNE Plot')
plt.show()
```

```python
perp_30_default_iter = tsne_df[
    (tsne_df['Dimension 1'] > 500) | (tsne_df['Dimension 1'] < -500) |
    (tsne_df['Dimension 2'] > 500) | (tsne_df['Dimension 2'] < -500)
]

perp_30_default_iter
```

## Perplexity 100

```python
# Drop the target column
features = df_uci.drop(columns=['default.payment.next.month'])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Perform t-SNE
tsne = TSNE(n_components=2, random_state=42, n_iter=3000, perplexity=80)

# Fit and transform the scaled features with a progress bar
tsne_result = []
with tqdm(total=len(scaled_features), desc="t-SNE Progress") as pbar:
    for i in range(0, len(scaled_features), 100):  # Batch processing for better performance
        batch_result = tsne.fit_transform(scaled_features[i:i+100])
        tsne_result.append(batch_result)
        pbar.update(len(batch_result))

# Concatenate the t-SNE results
tsne_result = np.concatenate(tsne_result, axis=0)

# Create a new DataFrame with the t-SNE results
tsne_df = pd.DataFrame(data=tsne_result, columns=['Dimension 1', 'Dimension 2'])

# Plot the t-SNE scatter plot
plt.scatter(tsne_df['Dimension 1'], tsne_df['Dimension 2'])
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('t-SNE Plot')
plt.show()
```

```python
# perp_80_iter3k = tsne_df[
#     (tsne_df['Dimension 1'] > 20) | (tsne_df['Dimension 1'] < -20) |
#     (tsne_df['Dimension 2'] > 20) | (tsne_df['Dimension 2'] < -18.5)
# ]

# perp_80_iter3k

perp_80_iter3k = tsne_df[
    (tsne_df['Dimension 1'] > 20) | (tsne_df['Dimension 1'] < -20) |
    (tsne_df['Dimension 2'] > 13) | (tsne_df['Dimension 2'] < -18.5)
]

perp_80_iter3k
```

```python
perp_80_iter3k_special = tsne_df[
    (tsne_df['Dimension 1'] > 10) & 
    (tsne_df['Dimension 2'] < -8.5) 
]

perp_80_iter3k_special
```

```python
df=pd.concat([perp_80_iter3k, perp_80_iter3k_special])
df
```

```python
tsne_outlier_plot = tsne_df.copy()

index_values = df.index.values

# Add a flag column to df1 and mark it as 1 if the index is present in df2, else 0
tsne_outlier_plot['Outlier'] = tsne_outlier_plot.index.isin(index_values).astype(int)
tsne_outlier_plot
```

```python
# Plot the t-SNE scatter plot
sns.scatterplot(data=tsne_outlier_plot, x= 'Dimension 1', y='Dimension 2', hue='Outlier')
#plt.xlabel('Dimension 1')
# plt.ylabel('Dimension 2')
# plt.title('t-SNE Plot')
plt.show()
```

```python
#Identifying real values of outliers to confirm outling nature
index_vals = df.index.values

df_uci_outliers = df_uci.copy()

df_uci_outliers['Outlier'] = df_uci_outliers.index.isin(index_vals).astype(int)
df_uci_outliers
```

```python
pay_list = ['PAY_0','PAY_2', 'PAY_3', 'PAY_4', 'PAY_5']
for col in pay_list:     
    sns.scatterplot(data=df_uci_outliers, x='LIMIT_BAL', y=col, hue='Outlier')
    plt.show()
```

Observation that the outliers appear to be grouped in the bands 2, 0, -1 and -2 for the PAY variables. 

```python
#look at proportion of default payments for outliers and non-outliers
proportional_counts = pd.crosstab(df_uci_outliers['default.payment.next.month'], df_uci_outliers['Outlier'], normalize=True)

print(proportional_counts)
```
