```python
import sklearn
import shap
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

# Apply test code to UCI data

```python
#Import dataset
df_uci = pd.read_csv(r'C:\Users\jordy\OneDrive\MSc_Python\Individual_Project\Data\UCI_Credit_Card.csv')
df_uci.head()
```

```python
df_uci.drop(columns=['ID'], inplace=True)
df_uci.head()
```

```python
X = df_uci.iloc[:,:-1]
y= np.array(df_uci.loc[:,'default.payment.next.month'])
# y=y.reshape((len(np.array(df_uci)),1))

# y=pd.DataFrame(y)
# y=y.squeeze
# print(X.shape)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
# Prepares a default instance of the random forest regressor
#model = RandomForestRegressor()
model = sklearn.ensemble.RandomForestRegressor(n_estimators=1000, max_depth=4)
# Fits the model on the data
model.fit(X_train, y_train)
```

```python
# Fits the explainer
explainer = shap.Explainer(model.predict, X_test)
# Calculates the SHAP values - It takes some time
uci_shap_values = explainer(X_test)
```

```python
shap.plots.bar(uci_shap_values)
```

```python
shap.plots.beeswarm(uci_shap_values)
```

```python
uci_shap_values
```

```python
df_export = pd.DataFrame()
df_export['UCI_SHAP'] = pd.DataFrame(uci_shap_values.values).abs().mean()
df_export.to_csv('UCI_SHAP.csv')
df_export
```
